```python
import json
import sqlite3
```

```python

path =  "Colab Notebooks/sqlite_taxi.db"
path_formated = fr"/content/drive/MyDrive/{path}"


connection = sqlite3.connect(path_formated)
cursor = connection.cursor()
QUERY = ("""
         CREATE TABLE IF NOT EXISTS TAXI(
          ID INTEGER,
          LATITUDE REAL,
          LONGITUDE REAL);
       """)



cursor.execute(QUERY)

json_file = "/content/drive/MyDrive/Colab Notebooks/taxi.json"
taxi_coordinates = json.load(open(json_file))['features'][0]['geometry']['coordinates']
print(f"List of taxi coordinates:\n{taxi_coordinates}")

ID = 1
for cord in taxi_coordinates:

    cord1 = (ID, cord[0], cord[1])
    
    cursor.execute('INSERT into TAXI values(?,?,?)', cord1)
    ID = ID + 1

print(f'\nTotal of {ID} data records inserted successfully.')
```

```python
QUERY2 = ("""
         SELECT AVG(LATITUDE), AVG(LONGITUDE) FROM TAXI;
       """)

result = cursor.execute(QUERY2)
result_fetchone = result.fetchone()
avg_lat = result_fetchone[0]
avg_long = result_fetchone[1]

print(f"avg_lat:{avg_lat}")
print(f"avg_long: {avg_long}")




```
